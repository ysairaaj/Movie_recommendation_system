```python
movieId_imdbId_ratings.info()
```

```python
movie_pivot = movieId_imdbId_ratings[:5002426].pivot_table(index= 'userId' , columns = 'original_title' ,values ='rating')
```

```python
movie_pivot.head(10)
```

```python
movieId_imdbId_ratings.head()
```

```python
ratings_count = movieId_imdbId_ratings[:5002426].groupby('original_title')['rating'].count()
```

```python
ratings_count.head()
```

```python
def correlation(x):
    df = pd.DataFrame(movie_pivot.corrwith(movie_pivot[x]) , columns = ['correlation'])
    df.dropna(inplace =True)
    a = df.join(ratings_count)
    a[a['rating'] >= 1000]
    return a[a['rating'] >= 1000].sort_values('correlation' ,ascending =False)
```

```python
correlation('Minions')
```

# Using KNN

```python
movie_features = movieId_imdbId_ratings[:5002426].pivot_table(index= 'original_title' , columns = 'userId' ,values ='rating').fillna(0)
```

```python
movie_features.head()
```

```python
from scipy.sparse import csr_matrix
csr_movie_matrix = csr_matrix(movie_features)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric = 'cosine')
model.fit(csr_movie_matrix)
```

```python
movie_features.loc['3 Idiots',:]
def knnfunc(x):
    indices = model.kneighbors([movie_features.loc[x,:]] , n_neighbors =100 ,return_distance =False)
    lst = [movie_features.index[x] for x in indices]
    a = pd.DataFrame()
    a['ranked'] = lst[0]
    return a
```

```python
knnfunc('3 Idiots')
```

# This function takes the ratio of correlation and knn to be used to get the final recommendation based on correlation and item based collaberative filtering (done by knnfunc())

```python
def final_recommendation(movie_name , cor = 1 , knn = 0) :
    df1 = correlation(movie_name).reset_index().reset_index()
    df2 = knnfunc(movie_name).reset_index()
    df3 = df1.merge(df2 , left_on = 'original_title' ,right_on = 'ranked' , how ='outer').drop(['correlation' ,'rating'] ,axis ='columns')
    df3['index_y'].fillna(len(df3.index) , inplace =True)
    df3['index_x'].fillna(len(df3.index) , inplace =True)
    #df3['original_title'].fillna('' , inplace =True)
    #df3['ranked'].fillna('' , inplace =True)
    #df3['title'] = df3['original_title'] + df3['ranked']
    df3['Rank'] = (df3['index_x']*knn + df3['index_y']*cor)/(knn + cor)
    df3.sort_values('Rank' ,ascending = True , inplace =True)
    df3['title'] = [x[1] if not pd.isna(x[1]) else x[2] for x in df3[['original_title','ranked']].itertuples()]
    return df3[['title' ,'Rank']][1:]
```

```python
final_recommendation('Harry Potter and the Chamber of Secrets' , cor = 0.5 , knn = 0.5)[:10]
```
